Nesse exercicio iremos carregar um arquivo .CSV e realizar filtros e transformações utilizando a biblioteca Pandas.



```python
# Importando as bibliotecas pandas
import pandas as pd
from google.colab import files #importando a biblioteca de upload de arquivos
```

```python
# Faz o upload do arquivo
uploaded = files.upload()

# Lê o arquivo CSV que foi carregado
for filename in uploaded.keys():
    df = pd.read_csv(filename)

print("Dados originais")
df.head()
```

```python
# Aplciar filtros (Exemplo : filtrar linhas onde a coluna 'idade' é maior que 30)

filtro_idade = df[df['idade']>30]
print("Dados filtrados (idade > 30):")
filtro_idade
```

```python
# Aplicando filtro pela coluna nome
filtro_nome = df[df['nome'] == 'Bob']
filtro_nome
```

```python
# Transformação de dados -> Alterando a idade de uma determinada pessoa

df.loc[df['nome'] == 'Bob', 'idade'] = 40
print('Alterando a idade do Bob')

df
```

```python
# Salvar o DataFrame alterado em um novo CSV

df.to_excel('Arquivo_alterado.xlsx', index = False)

# Baixar o arquivo
files.download('Arquivo_alterado.xlsx')
print('Arquivo Alterado e salvo como arquivo_alterado.xlsx')
```
