```python
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import requests


url = 'https://jsonplaceholder.typicode.com/users'

reponse = requests.get(url) #
data = reponse.json() # lendo dados no formato JSON


df = pd.json_normalize(data)
df.head()
```

```python
# Transformação de dados e tradução dos títulos das colunas

df.rename(columns={
    'id':'id_pessoa',
    'name':'nome',
    'username':'nome_de_usuario',
    'phone':'telefone',
    'website':'site',
    'address.street':'endereco',
    'address.suite':'numero_do_apartamento',
    'address.city':'cidade',
    'address.zipcode':'cep',
    'company.name':'nome_da_empresa',
    'company.catchPhrase':'frase_de_efeito'

}, inplace=True) #Inplace modifica o Data Frame original

df
```

```python
# Transformação de dados

df.loc[df['nome'] == 'Nicholas Runolfsdottir V', 'nome_da_empresa'] = 'Hoeger LLC'
df.loc[df['nome'] == 'Leanne Graham', 'nome_da_empresa'] = 'Hoeger LLC'
df.loc[df['nome'].isin(['Leanne Graham' , 'Nicholas Runolfsdottir V'])]
```

```python
# Filtrando dados

filtro_nome_empresa = df[df['nome_da_empresa'] == 'Hoeger LLC']
filtro_nome_empresa
```

```python
# Mostrando apenas colunas necessárias
df_modificado = df[['nome', 'nome_da_empresa', 'cidade', 'endereço', 'cep', 'telefone']]
df_modificado
```

```python
# Gerando arquivo excel
df_modificado.to_excel('fusão_de_empresas_filtrado.xlsx')
```
