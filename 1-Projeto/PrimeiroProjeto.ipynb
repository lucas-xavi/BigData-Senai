```python
# Importando as bibliotecas Pandas
import pandas as pd
from google.colab import files # Importando a biblioteca de upload de arquivos
```

```python
# Faz o upload do arquivo
uploaded = files.upload()
```

```python
# Lê o arquivo CSV que foi carregado
for filename in uploaded.keys():
    df = pd.read_csv(filename)

print("Dados originais")
df
```

```python
# Filtrando o DataFrame por Departamento
filtro_departamento = df[df['Departamento'] == 'T.i']
filtro_departamento
```

```python
# Ralizando transformação no cargo da pessoa selecionada
df.loc[df['Nome'] == 'Henrique', 'Cargo'] = 'Analista de Produção'
print("Alterando o cargo do Henrique")
df.loc[df['Nome'] == 'Henrique']
```

```python
# Gerando um novo arquivo atualizado

df.to_excel('Dados.xlsx', index=False)
files.download('Dados.xlsx')
```
